# Early Stopping

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.base import clone
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)

m, n = 1000, 1
X = np.random.randn(m, 1)
y = 3 * X**2 + .5 * X + 4 + (np.random.randn(m, 1) * 0.6)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
poly_scaler = Pipeline([
    ("poly_features", PolynomialFeatures(degree=90, include_bias=False)),
    ("std_scaler", StandardScaler())
]) 

X_train_poly_scaled = poly_scaler.fit_transform(X_train)
X_val_poly_scaled = poly_scaler.transform(X_val)

sgd_reg = SGDRegressor(
    max_iter=1, tol=.1, warm_start=True, penalty=None,
    learning_rate="constant", eta0=0.0005
)

minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled, y_train)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val, y_val_predict)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
```

### Summary of Learned Tricks & Techniques thus far ...

- Linear regression, Polynomial Regression
- The normal equation / The closed form - the optimal math equation
- Gradient Descent, Stochastic Gradient Descent, Mini-Batch Gradient Descent
    - metaheuristics to estimate the hypothesis $\theta$, not always optimal
- Learning curves
    - Cost function vs. Training set size
    - helps decide whether adding more instances would help
        - if not, use more complex model (e.g. increase the degree) or come up with better features
- Bias/Variance trade-off
- Regularization - is a way to contraint the model to reduce variance
    - Using regularization terms
        - Ridge (l2)
        - Lasso (l1)
        - Elastic Net
- **Early Stopping**
    - cost function vs. number of iterations / epoch
        - helps decide the best number of iteations
